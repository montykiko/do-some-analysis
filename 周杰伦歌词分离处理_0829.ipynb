```python
def df2mongo(df, coll):
    records = json.loads(df.T.to_json()).values()
    result = coll.insert_many(records)
    return result
```

```python
def send_request(text):
    try:
        response = requests.post(
            url="https://api.kismet-emotions.com/dev/emotionaggregate",
            headers={
                "Content-Type": "application/json; charset=utf-8",
                "Authorization": "Basic d2VpcWl0ZXN0OndlaXFpQVBJ",
            },
            data=json.dumps({
                "sentence": text,
                "emotionModel": 2
            })
        )
        json1_data = json.loads(response.content.decode("utf-8"))
        return json1_data
    except requests.exceptions.RequestException:
        return('HTTP Request failed')
```

```python
for ID in range(len(df)):
    Data = lyrics_jay.find_one({'ID':ID})
    data = send_request(Data['歌词'])
    data['ID'] = Data['ID']
    lyrics_new.insert_one(data)
```
