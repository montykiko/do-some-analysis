# 惠氏

```python
HS = pd.read_table('HS.txt')
```

```python
HS['date'] = pd.to_datetime(HS['评价时间(create_time)'])
HS.index = HS['date']
del HS['date']
```

```python
HS
```

## 惠氏各年度分离  #包含2013、2014、2015、2016、2017

```python
tfile = ["201%d" % i for i in range(3,8)]
```

```python
tfile
```

```python
for t in tfile:
    HS[t].to_csv('HS_%s.txt' % t,sep='\t',encoding='utf-8')
```

## 惠氏2015年12个月份

```python
filenames = ["2015-%02d" % i for i in range(1, 13)]
```

```python
filenames
```

```python
['HS_%s.txt' % i for i in filenames]
```

```python
for file in filenames:
    HS[file].to_csv('HS_%s.txt' % file.replace('-','_'),sep='\t',encoding='utf-8',index=False)
```

## 惠氏2016年12个月份

```python
filenames2 = ["2016-%02d" % i for i in range(1, 13)]
for file in filenames2:
    HS[file].to_csv('HS_%s.txt' % file.replace('-','_'),sep='\t',encoding='utf-8',index=False)
```
