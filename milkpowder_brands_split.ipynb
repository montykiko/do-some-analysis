```python
import pandas as pd
```

# 处理品牌

```python
df['brands'] = df['品牌'].str.findall(r'美素|惠氏|优诺能|雅培').str[0]
```

```python
df['brands'].replace(['惠氏','美素','雅培','优诺能'],['HS','MS','YP','YNN'],inplace=True)
```

```python
df
```

```python
filename = ['HS','MS','YP','YNN']
for file in filename:
    df.loc[df['brands']== file].to_csv('%s.txt' % file,sep='\t',encoding='utf-8',index=False)
```
