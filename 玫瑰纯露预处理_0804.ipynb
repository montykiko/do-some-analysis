```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
from datetime import datetime
import matplotlib as mpl
mpl.rcParams['font.sans-serif'] = ['SimHei']
import matplotlib.pyplot as plt
%matplotlib inline
```

# 接入吴华诚爬虫数据

```python
df = pd.read_csv('makeup.txt',sep ='\t')
df.shape
```

# 选择关注的字段

```python
data = df[['用户名(userInfo:name)','评价内容(commentInfo:content)','情绪极性','情绪浓度','情绪','关键词','评价时间(commentInfo:date)','店名(goodsInfo:name)']].copy()
data.head()
```

# 基本情绪

```python
basic_emo = data.loc[data['关键词']=='原句']
basic_emo.head()
```

```python
basic_emo['评价内容(commentInfo:content)'].isnull().value_counts()
```

## 打H、X标签

```python
basic_emo = basic_emo.sort_values('情绪极性',ascending = False).copy()
basic_emo.loc[basic_emo['情绪极性']>50,'label']='H'
basic_emo.loc[basic_emo['情绪极性']<=50,'label']='X'
basic_emo.head()
```

```python
names= set(basic_emo['店名(goodsInfo:name)'])
names
```

## 计算好、差评率

```python
cates = [len(basic_emo.loc[basic_emo['店名(goodsInfo:name)']==name]) for name in names]
b = [basic_emo.loc[basic_emo['店名(goodsInfo:name)']==name]['label'].value_counts()/cate*100 for name,cate in zip(names,cates)]
good_or_not = pd.DataFrame(b,index = names)
good_or_not
```

```python
good_or_not.to_csv('good_or_not.txt',sep='\t',encoding = 'utf-8')    #各品牌好差评率数据
```

# 关键词情绪

```python
key_emo = data.loc[~(data['关键词'] =='原句')]
key_emo.head()
```

```python
key_emo.shape
```

```python
key_emo.to_csv('key_emo.txt',sep = '\t',encoding ='utf-8',index = False)  #关键词对应情绪
```

# 读入程凯处理联想词json

```python
import json
with open('key_makeup.json','r',encoding = 'gbk') as file:
    f = file.read()
```

```python
data = json.loads(f)
data
```

```python
df = pd.DataFrame(data)
df = df.stack().reset_index()
df.columns = ['主题','关联词','value']
df.head()
```

```python
df['情绪浓度'] = df.value.str[0] / df.value.str[1]
df['频次'] = df.value.str[1]
df.head()
```

```python
df[['主题','关联词','情绪浓度','频次']].to_csv('words_link.txt',sep='\t',encoding = 'utf-8',index = False)   #联想词平均情绪浓度
```
