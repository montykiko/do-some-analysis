```python
import pandas as pd
from pandas import Series,DataFrame
import pymongo
import re,json
```

```python
client = pymongo.MongoClient('localhost',27017)
weixin = client['weixin']
half_life = weixin['half_life']
```

```python
df = pd.DataFrame(list(half_life.find()))
df.head()
```

```python
len(df)
```

```python
for i in range(len(df['article'])):
    print (df.iloc[i]['article'])
```

```python
df['new_title'] = df['title'].str.replace(r'[\/:*?"<>|]+','')
```

```python
df['new_title']
```

```python
for i in range(len(df['article'])):
    with open('%s.txt' % df.iloc[i]['new_title'] ,'w',encoding='utf-8') as f:
        f.write(df.iloc[i]['article'])
        f.close()
```
