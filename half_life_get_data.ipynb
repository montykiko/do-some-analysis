# 提取表情

```python
df['emo'] = df['content'].str.findall(r'\[(.*?)\]',re.I|re.M)
df['emo'] = df['emo'].str[0].replace(r'[\[\]]','\t')
df['emo'].map(lambda x:x.strip() if isinstance(x,str) else x)
df['emo'] = df['emo'].str.split('\t+')
df['emo'] = df['emo'].map(lambda x: ','.join(x) if isinstance(x, list) else '')
df['emo'].value_counts().head(10)
```

```python
df.head()
```

```python
df.groupby('category').polarity.mean().sort_values(ascending = False)
```

```python
df.groupby('category').polarity.std().sort_values(ascending = False)
```
