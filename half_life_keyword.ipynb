```python
with open (r'dfp.txt','r',encoding = 'utf-8') as file:
    words_dict = {}
    for line in file:
        if line =='':
            continue
        for char in to_remove:
            line = line.replace(char,' ')
            line = line.replace('  ',' ')
        seg_list = jieba.cut(line)
        refined_corpse = ' '.join(seg_list).split(' ')
        for w in refined_corpse:
            if w == '' or len(w) == 1:
                continue
            elif w in words_dict:
                words_dict[w]+=1
            else:
                words_dict[w] = 1
    sorted_dic = sorted(words_dict.items(),key=operator.itemgetter(1))
    sorted_dic.reverse()

writer = open(r'dfp_keywords.txt','w',encoding = 'utf-8')
for i in range(100):
    writer.write(sorted_dic[i][0]+'\t'+ str(sorted_dic[i][1])+'\n')
writer.close()          
```
